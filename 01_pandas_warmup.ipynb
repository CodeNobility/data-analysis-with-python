```python
# Step 8: Check unique values in categorical columns
dataset.select_dtypes(include=['float64']).nunique()
```

```python
# Step 9: Simple filtering examples
# All female passengers
len(dataset[dataset['Sex'] == 0])
```

```python
dataset['Sex'].value_counts()[0]
```

```python
dataset['Sex'].value_counts().rename(index={0: "Men", 1: "Women"})
```

```python
# Passengers older than 50
len(dataset[dataset['Age'] > 50])
```

```python
# Step 10: Sort passengers by Fare (descending)
dataset.sort_values(by='Fare', ascending=False)
```

```python
# Find the average age of male vs female passengers.

dataset.groupby('Sex')['Age'].mean().rename(index={0:'men', 1:'women'})
```

```python
# Find how many passengers paid fare above 100.

len(dataset[dataset['Fare'] > 100])
```
