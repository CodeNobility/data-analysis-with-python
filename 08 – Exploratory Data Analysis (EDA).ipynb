```python
# Step 1: Import Libraries and dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv("Datasets/Cleaned_dataset_titanic.csv")
```

```python
# Step 3: Survival Overview

dataset['survived'].value_counts(normalize = True)
```

```python
# Step 4: Survival by Gender

survived_by_gender = (
    dataset.groupby('sex')['survived']
    .mean()
    .reset_index()
)

sns.barplot(data = survived_by_gender, x='sex', y='survived')
plt.title('survived_by_gender')
plt.xlabel('sex')
plt.ylabel('survived')
plt.show()
```

```python
# Step 5: Survival by Passenger Class

sns.barplot(data = dataset, x = 'pclass', y = 'survived')
plt.title('survival by passenger class')
plt.xlabel('pclass')
plt.ylabel('survival rate')
plt.show()
```

```python
# Step 6: Age Analysis

sns.histplot(data = dataset, x = 'age', hue = 'survived', bins = 30, kde = True)
plt.title('age analysis')
plt.xlabel('age')
plt.ylabel('survival rate')
plt.show()
```

```python
# step 7: Fare Impact

sns.boxplot(data=dataset, x='survived', y ='fare')
plt.title("Fare vs Survival")
plt.show()
```

```python
#Step 8: Combined Insight (Class + Gender)

sns.barplot(data=dataset, x='pclass', y='survived', hue='sex')
plt.title("Survival by Class and Gender")
plt.show()
```

```python
 # Step 9: Correlation Check

sns.heatmap(dataset[['age', 'fare', 'sibsp', 'parch', 'survived']].corr(), annot=True)
plt.title("Correlation with Survival")
plt.show()
```
