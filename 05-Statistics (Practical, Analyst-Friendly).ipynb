```python
# import modules
import pandas as pd
import numpy as np
```

```python
# Step 1: Load Data
dataset = pd.read_csv("Datasets/Cleaned_dataset_titanic.csv")
```

```python
# overview of top 3 data
dataset.head(3)
```

```python
# Step 2: Mean vs Median (Very Important)
mean_age = dataset['age'].mean()
median_age = dataset['age'].median()

print('mean_age',mean_age)
print('median_age',median_age)
```

```python
# Step 3: Variance & Standard Deviation
var_age = dataset['age'].var()
std_age = dataset['age'].std()

print('var_age',var_age)
print('std_age',std_age)
```

```python
# Step 4: Distribution Visualization
import matplotlib.pyplot as plt

plt.hist(dataset['age'], bins = 20)
plt.title("age distribution")
plt.xlabel('age')
plt.ylabel('counts')
plt.show()
```

```python
# Step 5: Correlation Between Variables
corr = dataset[['age', 'fare']].corr()

print('corr',corr)
```

```python
# Step 6: Survival Rate by Group (Stat Thinking)

survived_rate = (
    dataset.groupby('sex')['survived']
    .mean()
    .reset_index()
)

print(survived_rate)
```

```python
# Step 7: Compare Two Groups (Simple Insight)
fare_by_pclass = (
    dataset.groupby('pclass')['fare']
    .mean()
    .reset_index()
)

print(fare_by_pclass)
```
