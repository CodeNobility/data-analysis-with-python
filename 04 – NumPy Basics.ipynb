```python
# import numpy
import numpy as np
```

```python
# Step 2: Create NumPy Arrays
arr = np.array([2, 4, 6, 8, 10])
print(arr)
```

```python
# checking type of array
print(type(arr))
```

```python
# Step 3: Create Arrays Quickly
zeros = np.zeros((2, 4))
ones = np.ones((3, 3))
range_array = np.arange(1, 10, 1).reshape(3,3)

print(zeros)
print(ones)
print(range_array)
```

```python
# Step 4: Indexing & Slicing
# first element
# last element
# slicing

print(arr[0])
print(arr[-1])
print(arr[2:4])
```

```python
# Step 5: Vectorized Operations (MOST IMPORTANT)

prices = np.array([100, 200, 300, 400])

# Apply 10% discount to all prices
discounted = prices * 0.9
print(discounted)
```

```python
# Step 6: Mathematical Operations

print(np.sum(prices))
print(np.mean(prices))
print(np.max(prices))
print(np.min(prices))
```

```python
# Step 7: Conditional Operations

high_prices = prices[prices > 250]
print(high_prices)
```

```python
# Step 8: Reshaping Arrays

matrix = np.arange(1, 7).reshape(2, 3)
print(matrix)
```

```python
# Step 9: NumPy with Pandas (Real Use Case)

import pandas as pd

df = pd.read_csv("Datasets/Cleaned_dataset_titanic.csv")

# Convert column to NumPy array
ages = df['age'].to_numpy()

print(np.mean(ages))
print(np.percentile(ages, 75))
```
