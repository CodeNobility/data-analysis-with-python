```python
# Step 1: Load Data

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_csv("Datasets/Cleaned_dataset_titanic.csv")
```

```python
# Step 2: Categorical vs Numerical (Bar + Hue)

# Question: Does gender affect survival differently across classes?

sns.barplot(data = dataset, x='pclass', y = 'survived', hue = 'sex')
plt.title('gender affect survival differently across classes')
plt.xlabel('pclass')
plt.ylabel('survived rate')
plt.show()
```

```python
# Step 3: Numerical vs Numerical (Scatter + Hue)
# Question: Does fare relate to survival?

sns.scatterplot(data = dataset, x = 'fare', y = 'age', hue = 'survived')
plt.title('Does fare relate to survival')
plt.xlabel('fare')
plt.ylabel('age')
plt.show()
```

```python
# Step 4: Distribution by Category (Boxplot)

# Question: Do survivors pay more?

sns.boxplot(data = dataset, x = 'survived', y = 'fare')
plt.title('Do survivors pay more')
plt.xlabel('survivors')
plt.ylabel('fare')
plt.show()
```

```python
# Step 5: Interaction Effect (Age Group + Survival)

def age_group(age):
    if age < 18:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

dataset['age_group'] = dataset['age'].apply(age_group)

sns.barplot(data = dataset, x = 'age_group', y = 'survived')
plt.title('Interaction Effect (Age Group + Survival)')
plt.xlabel('age distribution')
plt.ylabel('survival score')
plt.show()
```
