```python
# Goal
# Learn how to:
# Choose the right chart for the question
# Make visuals manager-friendly
# Avoid misleading or cluttered charts
# Dataset: Retail Transactions (same as Day 21)
```

```python
# Step 1: Start from the Question (Not the Chart)

# “Let’s make a chart”(Wrong)
# “What decision should this chart support?”(Right)

# Examples:

# “Which product category should we invest more in?”

# “Which customer segment drives revenue?”
```

```python
# Step 2: Choose the RIGHT Chart (Quick Mapping)
# Question	Best Chart
# Compare categories	Bar chart
# Trend over time	Line chart
# Distribution	Histogram / Box plot
# Relationship	Scatter plot
# Contribution	Bar (not pie)
```

```python
import pandas as pd
df = pd.read_csv("Datasets/retail_sales_dataset.csv")
```

```python
# Step 3: Improve an Existing Chart (Real Example)
# Bad Chart (Typical Beginner)
# No title
# No axis labels
# Too many colors


# Good Chart (Professional)
import seaborn as sns
import matplotlib.pyplot as plt

category_sales = (
    df.groupby('Product Category')['Total Amount']
      .sum()
      .sort_values(ascending=False)
)

sns.barplot(
    x=category_sales.values,
    y=category_sales.index
)
plt.title("Revenue by Product Category")
plt.xlabel("Total Revenue")
plt.ylabel("Product Category")
plt.show()
```

```python
# Step 4: Reduce Cognitive Load

# Remove anything that doesn’t add meaning:
# Gridlines (unless needed)
# Legends (if obvious)
# Extra colors

# Rule:
# If it doesn’t help the decision, remove it.
```

```python
# Step 5: Highlight What Matters
# Use annotation, not decoration.

# Prepare data
category_sales = (
    df.groupby('Product Category')['Total Amount']
      .sum()
      .sort_values(ascending=False)
)

# CREATE FIGURE (IMPORTANT)
plt.figure(figsize=(8, 5))

# Plot
sns.barplot(
    x=category_sales.values,
    y=category_sales.index
)

# Titles & labels
plt.title("Electronics Drives the Highest Revenue")
plt.xlabel("Total Revenue")
plt.ylabel("Product Category")

# Annotation (SAFE placement)
top_value = category_sales.iloc[0]
plt.text(
    top_value * 0.6,
    0,
    "Top category",
    va='center'
)

# SHOW explicitly
plt.tight_layout()
plt.show()
```

```python
# Step 6: Label with Insight, Not Description

# (Wrong)Title: “Revenue by Product Category”
# (Right)Title: “Electronics Drives the Highest Revenue”

# This alone makes you look senior.
```

```python
# Step 7: Common Visualization Mistakes 

# Using pie charts for comparison

# Using too many colors

# No units on axes

# Decorative charts without insight
```
