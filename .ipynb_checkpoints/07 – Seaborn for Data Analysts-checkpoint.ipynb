```python
# Step 1: Import Libraries and dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv("Datasets/Cleaned_dataset_titanic.csv")
```

```python
# Step 2: Count Plot – Survival by Gender
sns.countplot(data = dataset, x = 'sex', hue = 'survived')
plt.title('Survival by Gender')
plt.show()
```

```python
# Step 3: Box Plot – Age vs Survival

sns.boxplot(data=dataset, x='survived', y='age')
plt.title("Age Distribution by Survival")
plt.show()
```

```python
# Step 4: Box Plot – Fare by Class

sns.boxplot(data=dataset, x='pclass', y='fare')
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

```python
# Step 5: Heatmap – Correlation Matrix

corr = dataset[['age', 'fare']].corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Step 6: Pairplot – Multi-Variable EDA

sns.pairplot(dataset[['age', 'fare', 'survived']], hue='survived')
plt.show()
```

```python
# Heatmap with more numeric columns

numeric_df = dataset.select_dtypes(include=['int64', 'float64'])
numeric_df.columns
```

```python
# Heatmap with more numeric columns

corr_matrix = numeric_df.corr()
```

```python
# Heatmap with more numeric columns

plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm"
)
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```
