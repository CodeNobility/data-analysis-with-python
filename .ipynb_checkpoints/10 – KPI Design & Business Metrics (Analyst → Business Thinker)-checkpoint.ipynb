```python
# Step 1: What is a KPI? (Simple)

'''KPI (Key Performance Indicator) =
A number that tells how well something is performing.

Examples (business):
Conversion rate
Churn rate
Retention rate
Average order value

For Titanic (simulation):
Survival rate
Survival by segment
Risk groups'''
```

```python
# Step 2: Define Business Questions

'''Before metrics, ask:

Who is most likely to survive?

Which segment is at highest risk?

What factors improve survival chances?'''
```

```python
# Step 3: Core KPI – Overall Survival Rate

import pandas as pd

dataset = pd.read_csv("Datasets/Cleaned_dataset_titanic.csv")

overall_survival_rate = dataset['survived'].mean() * 100
print('overall_survival_percentage is',overall_survival_rate,'%')
```

```python
# Step 4: Segment KPIs (Very Important)
# Survival Rate by Gender

survival_by_gender = dataset.groupby('sex')['survived'].mean() * 100
print('survival rate by gender is \n',survival_by_gender)
```

```python
# Survival Rate by Passenger Class

survival_by_class = dataset.groupby('pclass')['survived'].mean() * 100
print('survival rate by pclass is',survival_by_class)
```

```python
# Step 5: Risk KPI (Reverse Metric)
# Risk = 1 − Survival Rate

risk_by_class = (1 - survival_by_class/100) * 100
risk_by_class
```

```python
# Step 6: Composite KPI (Advanced Thinking)
# Create a High-Risk Flag:

dataset['high_risk'] = (
    (dataset['sex'] == 'male') &
    (dataset['pclass'] == 3) &
    (dataset['age'] > 18)
)

print(round(dataset['high_risk'].mean() * 100),'%')
```

```python
# Step 7: KPI Table (Manager-Friendly)

kpi_table = (
    dataset.groupby(['sex', 'pclass'])['survived']
      .mean()
      .reset_index()
      .rename(columns={'survived': 'survival_rate'})
)

kpi_table
```
