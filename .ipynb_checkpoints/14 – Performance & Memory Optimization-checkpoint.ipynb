```python
# Step 1: Load Data

import pandas as pd
dataset = pd.read_csv("Datasets/Cleaned_dataset_titanic.csv")
```

```python
# Step 2: Check Memory Usage

dataset.info(memory_usage = 'deep')
```

```python
dataset.head(10)
```

```python
# Step 3: Optimize Data Types (BIGGEST WIN)
# Convert object → category

dataset['sex'] = dataset['sex'].astype('category')
dataset['embarked'] = dataset['embarked'].astype('category')
dataset['age_group'] = dataset['age_group'].astype('category')
```

```python
# Convert int → smaller int

dataset['sibsp'] = dataset['sibsp'].astype('int8')
dataset['parch'] = dataset['parch'].astype('int8')
dataset['survived'] = dataset['survived'].astype('int8')
dataset['pclass'] = dataset['pclass'].astype('int8')
```

```python
# Step 4: Use usecols While Reading CSV

df_small = pd.read_csv(
    "Datasets/Cleaned_dataset_titanic.csv",
    usecols=['survived', 'pclass', 'sex', 'age', 'fare']
)
```

```python
df_small
```

```python
# Step 5: Avoid apply(axis=1)
# avoid

dataset['family_size'] = dataset.apply(
    lambda r: r['sibsp'] + r['parch'] + 1,
    axis=1
) 

# Fast:
dataset['family_size'] = dataset['sibsp'] + dataset['parch'] + 1
```

```python
# Step 6: Use Vectorized Conditions

import numpy as np

dataset['fare_level'] = np.where(dataset['fare'] > 100, 'High', 'Low')
```

```python
# Step 7: Efficient Filtering

dataset.query("fare > 100 and pclass == 1")
```
