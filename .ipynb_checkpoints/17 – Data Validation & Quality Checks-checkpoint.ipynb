```python
# Step 1: Load & Inspect Data

import pandas as pd
dataset = pd.read_csv('Datasets/retail_sales_dataset.csv')
```

```python
# Step 2: Schema & Structure Checks

# Check:
# Column names
# Data types
# Missing values

dataset.info()
```

```python
dataset.head()
```

```python
# Step 2: Missing Value Checks

dataset.isnull().sum()
```

```python
df = dataset.dropna()
```

```python
# Step 3: Duplicate Checks

dataset['Transaction ID'].duplicated().sum()
```

If duplicates exist:
df = df.drop_duplicates(subset='Transaction ID')

```python
# Step 4: Range & Validity Checks (VERY IMPORTANT)

dataset[(dataset['Age'] < 0) | (dataset['Age'] > 100)]
```

```python
# Quantity validity
dataset[dataset['Quantity'] <= 0]
```

```python
# Price validity
dataset[dataset['Price per Unit'] <= 0]
```

```python
# Step 5: Logical Consistency Checks
# Total Amount check

dataset[dataset['Total Amount'] != dataset['Quantity'] * dataset['Price per Unit']]
```

```python
# Step 6: Outlier Detection
dataset['Total Amount'].describe()
```

```python
# optional
import seaborn as sns
sns.boxplot(x=dataset['Total Amount'])
```

```python
# Step 7: Categorical Validation
dataset_gender = dataset['Gender'].value_counts()
dataset_product = dataset['Product Category'].value_counts()
print(dataset_gender,"\n\n",dataset_product)
```

```python
# Step 8: Final Validation Checklist
assert dataset['Quantity'].min() > 0
assert dataset['Price per Unit'].min() > 0
assert dataset['Total Amount'].min() > 0
```
