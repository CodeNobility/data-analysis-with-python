```python
# Step 1: Import Libraries and dataset
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv("Datasets/Cleaned_dataset_titanic.csv")
```

```python

# Step 2: Bar Chart – Survival Rate by Gender
# Question: Who survived more?

survived_rate = (
    dataset.groupby('sex')['survived']
    .mean()
    .reset_index()
)

plt.bar(survived_rate['sex'], survived_rate['survived'])
plt.title('Survival Rate by Gender')
plt.xlabel('gender')
plt.ylabel('survived')
plt.show()
```

```python
# Step 3: Bar Chart – Survival Rate by Class

survived_rate_by_class = (
    dataset.groupby('pclass')['survived']
    .mean()
    .reset_index()
)
plt.bar(survived_rate_by_class['pclass'], survived_rate_by_class['survived'])
plt.title('Survival Rate by class')
plt.xlabel('class')
plt.ylabel('survived')
plt.show()
```

```python
# Step 4: Histogram – Age Distribution

# Question: How are passengers’ ages distributed?

plt.hist(dataset['age'], bins = 30)
plt.title('Age Distribution')
plt.xlabel('age')
plt.ylabel('count')
plt.show()
```

```python
# Step 5: Scatter Plot – Age vs Fare

# Question: Is there a relationship?

plt.scatter(dataset['age'], dataset['fare'])
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
# Step 6: Line Plot – Survival by Age (Binned)

dataset['age_bin'] = pd.cut(dataset['age'], bins=10)
survival_by_age = dataset.groupby('age_bin')['survived'].mean()

plt.plot(survival_by_age.values)
plt.title("Survival Rate Across Age Groups")
plt.xlabel("Age Groups")
plt.ylabel("Survival Rate")
plt.xticks(rotation=45)
plt.show()
```
