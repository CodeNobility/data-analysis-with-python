```python
# Segment Comparison (Category Ã— Gender)
# Question: Which categories perform better for which gender?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Datasets/retail_sales_dataset.csv')
seg = (
    df.groupby(['Product Category', 'Gender'])['Total Amount']
      .sum()
      .reset_index()
)

plt.figure(figsize=(8,5))
sns.barplot(data=seg, x='Total Amount', y='Product Category', hue='Gender')
plt.title("Revenue by Product Category and Gender")
plt.xlabel("Total Revenue")
plt.ylabel("Product Category")
plt.tight_layout()
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
# Extract Date Features

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Month_name'] = df['Date'].dt.month_name()
df['Day'] = df['Date'].dt.day
df['Day_name'] = df['Date'].dt.day_name()
```

```python
df.head(2)
```

```python
# Trend + Seasonality (Monthly)
# Question: Are there seasonal peaks?

monthly = (
    df.groupby(['Year','Month'])['Total Amount']
      .sum()
      .reset_index()
)

plt.figure(figsize=(8,4))
sns.lineplot(data=monthly, x='Month', y='Total Amount', hue='Year', marker='o')
plt.title("Monthly Revenue Trend by Year")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.tight_layout()
plt.show()
```

```python
# Outlier Detection (Customer Spend)
# Question: Are a few customers driving revenue?

cust = (
    df.groupby('Customer ID')['Total Amount']
      .sum()
      .reset_index()
)

plt.figure(figsize=(6,4))
sns.boxplot(x=cust['Total Amount'])
plt.title("Customer Total Spend Distribution")
plt.xlabel("Total Spend per Customer")
plt.tight_layout()
plt.show()
```

```python
# Relationship with Density (Quantity Ã— Price)
# Question: What combinations drive high value?

plt.figure(figsize=(6,5))
sns.scatterplot(
    data=df,
    x='Quantity',
    y='Price per Unit',
    size='Total Amount',
    sizes=(20,200),
    alpha=0.5
)
plt.title("Quantity vs Price (Bubble = Transaction Value)")
plt.tight_layout()
plt.show()
```

```python
# Focus the Viewer (Highlight, Donâ€™t Decorate)
# Question: Which category should we prioritize?

cat = (
    df.groupby('Product Category')['Total Amount']
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(7,4))
sns.barplot(x=cat.values, y=cat.index, color='lightgray')
sns.barplot(x=[cat.iloc[0]], y=[cat.index[0]], color='steelblue')
plt.title("Electronics Is the Primary Revenue Driver")
plt.xlabel("Total Revenue")
plt.ylabel("Product Category")
plt.tight_layout()
plt.show()
```
