```python
# After (Dashboard-Ready)
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('Datasets/retail_sales_dataset.csv')
cat = (
    df.groupby('Product Category')['Total Amount']
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(7,4))
sns.barplot(x=cat.values, y=cat.index, color='steelblue')

plt.title("Electronics Is the Primary Revenue Driver")
plt.xlabel("Total Revenue")
plt.ylabel("")

plt.tight_layout()
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Extract Date Features

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Month_name'] = df['Date'].dt.month_name()
df['Day'] = df['Date'].dt.day
df['Day_name'] = df['Date'].dt.day_name()
```

```python
# Step 3: Refactor Chart #2 (Monthly Trend)
# Clean Trend Chart

monthly = (
    df.groupby(['Year','Month'])['Total Amount']
      .sum()
      .reset_index()
)

plt.figure(figsize=(7,4))
sns.lineplot(data=monthly, x='Month', y='Total Amount', marker='o')

plt.title("Monthly Revenue Shows Seasonal Peaks")
plt.xlabel("Month")
plt.ylabel("Total Revenue")

plt.tight_layout()
plt.show()
```
